```python
from tensorflow.keras.preprocessing.image import img_to_array
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import relu,linear
from tensorflow.keras.losses import CategoricalCrossentropy , mae,mse
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imutils import paths
import numpy as np
import cv2
import os
import pandas
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df=pandas.read_csv("train.csv")
df
```

```python
df=pandas.concat([df,df[df.TravelInsurance=="Yes"] ],ignore_index=True) 
```

```python
df[df.TravelInsurance=="Yes"]
```

```python
string_columns=df.select_dtypes(include=['object']).columns.tolist()
string_columns.remove("Customer Id")
string_columns, len(string_columns)
```

```python
label_encoders=[]
for str_col in string_columns:
    label_encoder = LabelEncoder()
    df[str_col] = label_encoder.fit_transform(df[str_col])
    label_encoders.append(label_encoder)
```

```python
X=df.drop(['Customer Id',"TravelInsurance"], axis=1).to_numpy()
X.shape
```

```python
# Normalize each feature using its own MinMaxScaler
X_normalized = np.empty_like(X, dtype=float)
scalers = []
for feature_index in range(X.shape[1]):
    scaler = StandardScaler()
    feature = X[:, feature_index].reshape(-1, 1)
    scaler.fit(feature)
    X_normalized[:, feature_index] = scaler.transform(feature).flatten()
    scalers.append(scaler)

print("Original Matrix X:")
print(X)
print("\nNormalized Matrix X:")
print(X_normalized)

# for feature_index, scaler in enumerate(scalers):
#     print(f"Scaler for feature {feature_index}:", scaler)
```

```python
y=df.TravelInsurance.to_numpy()
y.shape
```

```python
trainX, crossX, trainY, crossY = train_test_split(X_normalized,y, test_size=0.10,random_state=42)
```

```python
model=Sequential([
    Input( shape= (X.shape[1],) ),
    Dense(256,activation="relu"),
     Dense(128,activation="relu"),
    BatchNormalization(),
    Dense(64,activation="relu"),
    Dense(32,activation="relu"),
    Dense(32,activation="relu"),
    Dense(1,activation="sigmoid")
])
model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss="binary_crossentropy",metrics=["accuracy"])
```

```python
save_path="./result.hdf5"
checkpoint = ModelCheckpoint(save_path, monitor="val_loss",
	save_best_only=True, verbose=1)
callbacks = [checkpoint]
model.fit(trainX,trainY,epochs=500
          ,validation_data=(crossX, crossY)
          ,callbacks=callbacks,batch_size=32
         )
```

```python
loaded_model=tf.keras.models.load_model(save_path)
```
